# hf_model_evaluator — Test Environment

## 1. Single model — NLI

### `NLIEvaluator(base_dir)`

| Parameter | Type | Default | Description |
|-----------|------|---------|-------------|
| `base_dir` | str | `'./results'` | Directory where results and logs are saved. Relative paths are resolved to absolute at instantiation. |

### `evaluator.evaluate(model_id)`

| Parameter | Type | Description |
|-----------|------|-------------|
| `model_id` | str | HuggingFace model ID. Must have pipeline tag `zero-shot-classification` or `text-classification`. |

```python
from qlatent.hf_model_evaluator import NLIEvaluator

evaluator = NLIEvaluator(base_dir='./results')
evaluator.evaluate('typeform/distilbert-base-uncased-mnli')
```

## 2. Batch controller — NLI

### `evaluator.run(chunk_file, only_zero_shot, hf_token)`

| Parameter | Type | Default | Description |
|-----------|------|---------|-------------|
| `chunk_file` | str | `None` | Path to a text file with one model ID per line. If omitted, all HuggingFace NLI models are fetched automatically. |
| `only_zero_shot` | bool | `False` | Restrict to `zero-shot-classification` models only. Ignored when `chunk_file` is provided. |
| `hf_token` | str | `None` | HuggingFace API token. Reduces rate-limiting. Falls back to `HF_TOKEN` environment variable if set. |

> **`only_zero_shot=False` (default):** fetches both `zero-shot-classification` and `text-classification` models from HuggingFace. This covers a much larger set of models but will take significantly longer to run.
>
> **`only_zero_shot=True`:** fetches only `zero-shot-classification` models, which are explicitly trained for NLI inference. Recommended for faster targeted runs.

```python
with open('./models.txt', 'w') as f:
    f.write('typeform/distilbert-base-uncased-mnli\n')
    f.write('facebook/bart-large-mnli\n')

evaluator.run(chunk_file='./models.txt')
```

## 3. Full automated pipeline — NLI

Fetches all zero-shot NLI models from HuggingFace automatically and runs the full pipeline.

```python
evaluator.run(only_zero_shot=True)
```

## 4. Inspect NLI results

```python
import pandas as pd

df = pd.read_csv('./results/QMNLI/ASI.csv', encoding='utf-8-sig')
df.head()
```
